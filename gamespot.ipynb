```python
import os
from dotenv import load_dotenv
from requests import get
```

```python
url = 'http://www.gamespot.com/api/games/?'
API_KEY = os.getenv('GAMESPOT_API_KEY')
```

```python
params = {
    'api_key': API_KEY,
    'format': 'json'
}

# headers = {
#     'content-type': 'application/xml; charset=utf-8'
# }
```

```python
res = get(url, params=params).json().get('results')
print(res)
```

```python
for game in res:
    print(game.get('name'))
    print(game.get('themes'))
    print(game.get('genres'))
    print(game.get('release_date'))
    print(game.get('description'))
    print('---------------------')
```
