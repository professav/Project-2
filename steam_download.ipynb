```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(algorithm='brute')
model.fit(sparse_games)
```

```python
from typing import List
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors

def get_game_vectors(df: pd.DataFrame, game_titles: List[str]) -> np.ndarray:
    game_vectors = []
    for title in game_titles:
        if title in df.index:
            game_vectors.append(df.loc[title].values)
        else:
            raise ValueError(f"The book '{title}' is not in the DataFrame index.")
    return np.array(game_vectors)

def books_recommendation(user_liked_games: List[str], model: NearestNeighbors, df: pd.DataFrame, n_neighbors: int = 8) -> List[str]:
    game_vectors = get_game_vectors(df, user_liked_games)
    average_vector = np.mean(game_vectors, axis=0).reshape(1, -1)
    distance, suggestion = model.kneighbors(average_vector, n_neighbors=n_neighbors + len(user_liked_games))
    
    suggested_books = [df.index[suggestion[0][i]] for i in range(len(suggestion[0]))]
    
    filtered_suggestions = [book for book in suggested_books if book not in user_liked_games][:n_neighbors]
    
    return filtered_suggestions

user_liked_games = [
    'SOS',
    'House Flipper',
    'DYNASTY WARRIORS 9',
]

suggested_books = books_recommendation(user_liked_games, model, game_pivot, 9)
for book in suggested_books:
    print(book.capitalize())
```
